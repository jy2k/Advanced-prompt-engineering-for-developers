```python
original_question = """ 
    How many protons, neutrons, and electrons does potassium-40 have when it is part of K2SO4?
    Protons: 
    Neutrons:
    Electrons: 
"""
print("[red]"+llm.invoke(original_question)+"[red]")

step_back_question = """
    What are the principles behind the question?
    Question:
"""

stepback_answer = llm.invoke(step_back_question + original_question)
print(stepback_answer)

final_answer = llm.invoke(original_question + stepback_answer)
print("[green]"+final_answer+"[green]")
```

```python
original_question = """ 
    What happens to the pressure of an ideal gas when temperature increases 
    by a factor of 2, and volume increases by a factor of 8?
"""
print("[red]"+llm.invoke(original_question)+"[red]")

step_back_question = """
    What are the fundamental concepts or principles involved in this problem?
"""

stepback_answer = llm.invoke(step_back_question + original_question)
print(stepback_answer)

final_answer = llm.invoke("Use: " + stepback_answer + " and asnwer question: " + original_question + " answer:")

print("[green]"+final_answer+"[green]")
```
